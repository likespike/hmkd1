#### 과제2_0421.
위의 그래프를 x,y 구간을 400으로 조밀하게 변경하여 다시 표현하세요.

```python
def f(x):
    return x**3 - 3*x**2 + x

x = np.linspace(-1,3,400)
y = f(x)
```

```python
# Q. f(x)를 그래프로 표현하세요.

import matplotlib as mpl
# from matplotlib import rc
import matplotlib.pylab as plt

mpl.rc('font', family='Malgun Gothic')
mpl.rcParams['axes.unicode_minus'] = False    # 마이너스 단위 처리

plt.figure(figsize=(10,6))
plt.plot(x, y, 'g')    # marker 빼줌 ('ro' -> 'r')
plt.xlim(-2,4)    # 범위 조절
plt.title('함수 $(f(x) = x^3 = 3x^2 + x$의 그래프')
plt.xlabel('x', fontdict=({'fontsize':14, 'color':'red'}))
plt.ylabel('y', fontdict=({'fontsize':14, 'color':'red'}), rotation=0)
plt.xticks(np.arange(-1, 4))    # 눈금 조절
plt.yticks(np.arange(-5, 4))
plt.show()
```

#### 과제3_0421.
다음 함수의 역함수를 구하고 그래프로 표현하세요.

```python
def f(x):
    return(x**2 + 1)
```

```python
def f(x):
    return(x**2 + 1)

def finv(x):
    return np.sqrt(x-1)
    
x = np.linspace(0,3,300)
plt.figure(figsize=(6,6))

plt.plot(x, f(x))
plt.plot(x,x, '--')
plt.plot(x, finv(x))
plt.xlim(0,3)
plt.ylim(0,3)
plt.grid(True)
```

#### 과제4_0421.
로그함수를 시각화하여 표현하세요.

```python
xx = np.linspace(-2,2,100)
yy = np.log(xx)

plt.figure(figsize=(10,6))

plt.plot(xx,yy)
plt.axhline(1,c='r',ls='--')
plt.axhline(0,c='r',ls='--')
plt.axvline(1,c='r',ls='--')

plt.title('로그함수')
plt.xlabel('$x$')
plt.ylabel('$log(xx)$')
plt.show()
```

#### 과제6_0421. mt1 행렬에 대하여 다음을 수행하세요.
