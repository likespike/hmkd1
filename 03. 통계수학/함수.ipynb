```python
# linspace - linearly spaced, 주어진 범위 내에서 선형적으로 분포된 일정한 간격의 값을 생성


def f(x):
    return x**3 - 3*x**2 + x

x = np.linspace(-1,3,9)
y = f(x)
```

# 역함수
- 어떤 함수의 입력/출력 관계와 정반대의 입출력 관계를 갖는 함수
- 원래의 함수에서 x축과 y축이 바뀐 것이므로 y=x가 나타내는 직선을 대칭축으로 대칭인 함수의 그래프

```python
# $ 기호는 LaTeX 수학 표현식을 표시하기 위한 기호입니다.
# Matplotlib에서는 $ 기호를 사용하여 수학 표현식을 표시할 수 있습니다. 

# ple.axiw('equal') - Matplotlib에서 그래프의 x축과 y축의 눈금 간격을 같게 설정하는 함수

def f1(x):
    return x ** 2

def f1inv(x):
    return np.sqrt(x)

x = np.linspace(0,3,300)    # 끝점의 숫자도 포함함 (0, 0.01, ..., 2.99, 3)
# plt.figure(figsize=(10,6))
plt.plot(x,f1(x),'r', label='함수 $f(x) = x^2$')    # 실제 연산자라서 **는 못 쓰는듯 <-> sympy에서는 가능함
plt.plot(x,f1inv(x), 'b', label='역함수 $f^{-1}(x) = \sqrt{x}$')
plt.plot(x,x, 'g--')
plt.axis('equal')    # x축과 x축의 눈금 간격을 같게 설정함
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()
plt.title('역함수의 그래프')
plt.show()
```

- 바뀐 함수를 y에 대해서 푼 후, x로 정리합니다

#### 지수함수
- 정해진 어떤 값을 입력값으로 거듭제곱하여 출력을 구하는 함수
- 거듭제곱을 할 때 아래에 있는 수를 밑이라고 하며 밑을 오일러 수 e(약 2.718)로 하여 거듭제곱하는 함수를 지수함수라고 한다.
- exp는 자연 상수 e의 거듭제곱을 나타내는 함수
- 자연 상수 e는 대략 2.71828의 값을 가지는 상수로, 자연계에서 매우 많은 현상을 모델링하는 데에 이용됩니다.
- y = exp(x)는 y가 e의 x 거듭제곱값과 같다는 의미입니다. 예를 들어, x=2인 경우, y는 e의 2제곱인 약 7.389에 가까운 값이 됩니다.
- $y=e^x$, y=exp(x) = exp x 로 표시
- 로그함수의 밑으로 자연로그 함수(ln)을 사용하는데 이때 밑이 오일러 상수 e이다.
- 자연상수e를 밑으로 하는 지수함수를 미분하면 그대로 자연상수e를 밑으로 하는 지수함수 나온다. ex) $d(e^x)/dx = e^x$
- 우리가 쓰는 정규분포 같은 많은 포뮬러들이 다 e를 밑으로 하는 지수함수가 기반이다

#### 미분
- 변화 예측
- 변화 유발자의 대표선수 : 시간
- 시간이 가면 대나무가 큰다
- 대나무 키는 갑자기 크거나 작아지지 않는다 -> 대나무 키는 부드럽게 이어져 있다
- 그래서 중요한 것은 바로 직전 하루 저 대나무가 자란 키
- 변화를 좁혀서 좁혀서 꺼내면 예측이 더 정확해진다
- 순간의 변화를 끄집어 내는건 : d한다라고 한다
- 그 무수히 많은 순간의 변화들을 다 합하는건 : 인테그랄 
- 연속적이지 않을 때는 d는 델타, 인테그랄은 시그마로 표현한다
- 미분을 하면 차원이 줄어든다
- 계속 변하는 것을 x로 쓴다
- x가 변하면 전체 갯수가 변한다 -> 변화 유발자
