#### python random 
1. random.random() : 0.0에서부터 1.0 사이의 실수(float)를 반환
2. random.uniform(a, b) : 인자로 들어온 a~b 사이의 실수(float)를 반환
3. randint(a, b) : 인자로 들어온 a, b 사이의 랜덤한 정수(int)를 반환
4. randrange(a, b), randrange(b) :  a <= x < b 의 범위 내에서의 랜덤한 정수(int)를 반환
                                    randrange(b)는 0 <= x < b 의 범위 내에서의 랜덤한 정수(int)를 반환
5. random.choice(seq) : 인자로 받은 리스트, 튜플, 문자열, range 에서 무작위로 하나의 원소 반환
6. random.sample(seq or set, N) : 첫번째 인자로 받은 시퀀스데이터 or set에서 N개의 랜덤하고, unique하고, 순서상관없이 인자를 뽑아서 리스트로 만들어서 반환
7. random.shuffle(seq) : 데이터의 순서를 무작위로 랜덤하게 바꾸어 주는 함수

```python
import random 

menu = ["짬뽕","분식집","맥도날드","쉑쉑버거","편의점","닭곰탕"]
#메뉴에 추가하는 코드
print(f"기본으로 제공하는 메뉴 \n{menu} \n입니다.\n")
while 1:
    a = input("주변 음식점 이름을 적으세요. 없으면(plzstp) \n")
    b = a.strip().lower()
    if b == "plzstp":
        break
    else:
        menu.append(a)
random.shuffle(menu)
print("\n",menu)
# 추천수를 적고 메뉴를 찍어주는 코드 
while 1:
    try:
        num = int(input("추천받고 싶은 갯수를 말하세요. \n"))
        pick = random.sample(menu,num) ; print(f"\n추천 메뉴는 {pick} 입니다.\n \n")
        break
    except:
        print("숫자를 적어주세요.")

choice = input("마음에 들지 않으신가요? (y or n)").strip().lower()  
#마음에 들지 않을때 다시 생각하게 해주는 코드
if choice == "y":
    recmd_num = random.randrange(1,len(menu)+1)
    user_choice = input(f"\n제가 추천하고 싶은 \n {menu} \n에 메뉴 숫자는 {recmd_num}입니다. 이대로 추천할까요? (y,n) \n").strip().lower()
    if user_choice == "y":
        re_pick = random.sample(menu,recmd_num) ; print(f"추천 메뉴는 {re_pick} 입니다.")
    elif user_choice == "n":
        print("\n식사맛있게하십시오 ^^ ")
    else: 
        print("\n숫자가 아닙니다. 프로그램을 종료합니다.")
elif choice == "n":
    print("\n식사맛있게하십시오 ^^")
else:
    print("\n잘못 입력하셨습니다. 프로그램을 종료합니다.")
```

```python
# Q. 반복문으로 팩토리얼을 구하는 사용자 함수를 작성하고 5!, 10!을 구하세요.

def factorial(n):
    output = 1
    for i in range(1,n+1):
        output *= i
    return output

print(factorial(5))
print(factorial(10))
```

```python
# Q. 재귀란 '자기자신을 호출하는 것'을 의미한다. 재귀함수를 이용해서 팩토리얼을 구하는 사용자 함수를
# 작성하고 1!, 5!, 10!를 구하세요.
# factorial(n) = n * factorial(n-1) (n >= 1)

def factorial(n):
    if n ==0:
        return 1
    else:
        return n*factorial(n-1)
    
print(factorial(1))
print(factorial(5))
print(factorial(10))
```

```python
#. 원의 면적과 둘레의 길이를 구하는 사용자 함수를 작성하여 반지름이 5와 10일 때 면적과 둘레를 구하세요

pi = 3.141592

def circle(r):
    a = pi * r * r
    b = 2 * pi * r
    print(f'반지름 {r}에 대한 원의 면적은 {round(a,2)}이고, 둘레는 {round(b,2)}입니다.')

circle(5)

# print(circle(10))
```

```python
# Q. input("문자 수정을 원하시면 '수정'을 입력하시고, 문자 개수를 착기 및 삭제를 원하시면 '찾기'를 입력해 주세요")
# 아래 시나리오가 구현될 수 있도록 코드를 작성하세요.
# replace() # find()


while 1:
    i = input("문자 수정을 원하시면 '수정'을 입력하시고, 문자 개수를 찾기 및 삭제를 원하시면 '찾기'를 입력해 주세요(종료:q)> ")
    print()
    
    if i == '수정':
        while 1:
            sentence = input('글을 입력해주세요>(종료:q) ')
            print()
            if sentence == 'q':
                print("종료합니다.")
                break
                
            elif sentence not in "":            
                print("입력하신 글자는 다음과 같습니다. ====>  ", sentence)
                print()

                while 1:
                    e = input('글에서 수정하고 싶은 문자를 입력해주세요(종료:q)> ') # while문 밖으로 나가면 안 됨
                    print()
                    
                    if e == 'q':
                        print("종료합니다.")
                        break
                    elif e == "":
                        print("아무 것도 입력하지 않았습니다.")
                    elif e not in sentence:
                        print("수정하려는 문자가 없습니다. 다시 입력해 주세요.")
                    else:
                        while 1:
                            e1 = input("문자를 어떻게 수정하시겠습니까?(종료:q)> ")
                            print()
                            
                            if e1 == 'q':
                                print("종료합니다.")
                                break
                            if e1 not in "": 
                                result = sentence.replace(e, e1)
                                print("수정하신 문장은 다음과 같습니다. ====>", result)
                                break                                    
                            else:
                                print("아무 것도 입력하지 않았습니다.")
                        break
                break # elif의 break임, 위의 if에 break가 있는 것처럼 -> 이게 상위 while문 멈추게 함
        
            else:
                print("아무 것도 입력하시지 않았습니다.")
                
        break # 요거 중요, 안에 break 해주었으면, 밖에 while도 종료할 수 있도록 break해주어야 함
            
    elif i == '찾기':
        while 1:
            sentence = input('글을 입력해주세요>(종료:q) ')
            print()
            if sentence == 'q':
                print("종료합니다.")
                break
                
            elif sentence not in "":            
                print("입력하신 글자는 다음과 같습니다. ====>  ", sentence)
                print()

                while 1:
                    e = input('글에서 찾고 싶은 문자를 입력해주세요(종료:q)> ') # while문 밖으로 나가면 안 됨
                    print()
                    
                    if e == 'q':
                        print("종료합니다.")
                        break
                    elif e == "":
                        print("아무 것도 입력하지 않았습니다.")
                    elif e not in sentence:
                        print("찾으려는 문자가 없습니다. 다시 입력해 주세요.")
                    else:
                        print(sentence)
                        c = sentence.count(e)
                        print(f'해당 글에 존재하는 입력하신 문자의 개수는 {c}개 입니다.')
                        while 1:
                            e1 = input('해당 글에서 입력하신 문자를 모두 삭제하시겠습니까?(종료:q)(Y/N)> ')
                            print()
                            
                            if e1 == 'q':
                                print("종료합니다.")
                                break
                            elif e1 == "":
                                print("아무 것도 입력하지 않았습니다.")
                            elif e1.upper() == 'Y':
                                result = sentence.replace(e, "")
                                print("삭제하신 문장은 다음과 같습니다.\n", '====>', result)
                                break
                            elif e1.upper() == 'N':
                                print("알겠습니다. 프로그램을 종료합니다.")
                                break
                                
                            else: # e1.upper() != "Y" or e1 != "N":
                                print("잘못 입력하셨습니다. 다시 입력해 주십시오.")
                                   
                        break
                break # elif의 break임, 위의 if에 break가 있는 것처럼 -> 이게 상위 while문 멈추게 함
        
            else:
                print("아무 것도 입력하시지 않았습니다.")    
        
        break # 조건 문 안으로 들어가는듯 , 들여쓰기        
              
    elif i == 'q':
        print("종료합니다. 이용해 주셔서 감사합니다!")
        break       
        
    else:
        print("잘못 입력하셨습니다.")
    
```

```python
# A.
# strip() 함수를 사용하면 주어진 문자열에서 양쪽 끝에 있는 공백고 \n 기호를 삭제
# re.sub() 정규표현식 문자열 치환
# title() 함수 첫글자를 대문자로 변환

import re

states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']

def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]', '', value)
        value = value.title()
        result.append(value)
    return result

clean_strings(states)
```

```python
# 나영
# 문자열 리스트 states를 정형화하여 아래와 같이 출력하세요
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda','south   carolina##', 'West virginia?']

#양측 공백 제거
#첫글자만 대문자, 나머지 소문자 (공백 이후 글자는 대문자)
#특수문자 제거

for i in range(len(states)): # 7번 돌림
    a=[]
    states[i]=states[i].strip(' ')
    states[i]=states[i].lower()
    states[i]=states[i][0].upper()+states[i][1:]
    
    for j in range(1,len(states[i])):
        if states[i][j-1]==' ':
            states[i]=states[i].replace(states[i][j],states[i][j].upper())
        
        if not states[i][j].isalnum() and states[i][j]!=' ':
            a.append(states[i][j])
            
    for k in states[i]:
        if k in a:
            states[i]=states[i].replace(k,'')

print('\n'.join(states))
```

```python
def getTextFreq(filename):
    with open(filename,'r') as f:
        text = f.read()
        fa = {}
        for c in text:
            if c in fa:
                fa[c] +=1
            else:
                fa[c] = 1
    return fa

ret = getTextFreq('mydata.txt')
ret = sorted(ret.items(),key=lambda x:x[1], reverse=True)
print(ret)
for c, freq in ret: # key, value를 각각 받아 올 수 있음
    if c == '\n':
        continue
    print('[%c] -> [%d]회 나타남' %(c,freq))
```

파이썬 if __name__ == "__main__" 의미

- 메인 함수의 선언, 시작을 의미합니다. 해당 코드 밑에 main 등의 함수 호출 코드를 작성해서 함수의 기능을 수행
- 직접 modex2.py 파일을 실행하는 경우에는 __name___ 변수에 __main__ 이라는 값이 할당되고 이 조건절이 참이 되어 if문 다음 문장들이 수행
- 모듈을 불러와서 사용하는 경우 __name__ 변수에는 모듈 이름이 저장되고 이 조건절이 거짓이 되어 if문 다음 문장들이 수행되지 않음
- if __name__ == "__main__"와 같이 조건문을 사용하여 터미널에서와 같이 직접 호출되어 사용될 때는 그 자체로 기능을 수행하고, 동시에 다른 모듈에서 필요한 함수 등을 제공

파이썬 프로그램은 기본적으로 하나의 쓰레드(Single Thread)에서 실행된다. 즉, 하나의 메인 쓰레드가 파이썬 코드를 순차적으로 실행한다. 코드를 병렬로 실행하기 위해서는 별도의 쓰레드(Subthread)를 생성해야 하는데, 파이썬에서 쓰레드를 생성하기 위해서는 threading 모듈 (High 레벨) 혹은 thread 모듈 (Low 레벨)을 사용할 수 있다. 일반적으로 쓰레드 처리를 위해서는 thread 모듈 위에서 구현된 threading 모듈을 사용하고 있으며, thread 모듈은 (deprecate 되어) 거의 사용하고 있지 않다.

```python
# long_task 함수는 수행하는 데 5초의 시간이 걸리는 함수이다. 위 프로그램은 이 함수를 총 5번 반복해서 
# 수행하는 프로그램이다. 
# 이 프로그램은 5초가 5번 반복되니 총 25초의 시간이 걸린다.

import time

def long_task(): # 5초의 시간이 걸리는 함수
    for i in range(5):
        time.sleep(1)
        print("working:%s\n" % i)
    
print("Start")

for i in range(5):
    long_task()
    
print("End")
```

```python
# 파이썬에서 쓰레드를 실행하기 위해서는, threading 모듈의 threading.Thread() 함수를 호출하여 
# Thread 객체를 얻은 후 Thread 객체의 start() 메서드를 호출
# 쓰레드가 실행할 함수 (혹은 메서드)를 작성하고 그 함수명을 threading.Thread() 함수의 target 아큐먼트에 지정
# 프로그램을 수정하고 실행해 보면 25초 걸리던 작업이 5초 정도에 수행되는 것을 확인할 수 있다. 
# threading.Thread를 사용하여 만든 스레드 객체가 동시 작업을 가능하게 해 주기 때문임

# long_tstk()를 동시에 실행해 준다(한 process에서)

import time
import threading

def ling_task():
    for i in range(5):
        time.sleep(1)
        print("working:%s\n" % i)
        
print("Start")

threads = []
for i in range(5):
    t = threading.Thread(target=long_task)
    threads.append(t)
    
for t in threads:
    t.start()
    
for t in threads:
    t.join() # join으로 스레드가 종료될 때까지 기다린다. 중구난방안되게?? 이거 없으면 바로 End가 출력된다
    
print("End")
```

```python
# [과제] 123456789를 입력하면 천 단위로 콤마로 구분하여 출력하세요.

i = input('123456789를 입력해 주세요.') #하나하나가 str임

result=[] # append() 사용을 위해 빈 리스트 변수 생성

for j in i:
    if eval(j) %3 == 0 and eval(j) //3 != 3: # eval로 str을 int로 변경, 그리고 몫이 3이 아닌 것으로 9 뒤에 쉼표 방지
        result += j # 3, 6 숫자 다음 ',' 추가를 위해 result에 먼저 j값을 더해 줌 
        result.append(',') 
    else:
        result += j
        
result1 = "" #자료형 변경을 위해 빈 문자열 변수 생성

for k in result: # 자료형을 리스트 -> 문자열로 변경
    result1 += k 
                
print(result1)
```

과제 - Q56.
파일이름과 특정 단어를 인자로 입력하면 그 파일에 있는 특정 단어의 개수를 반환하는 사용자 함수를 이용하여 mydata 파일에서 how의 개수를 출력하세요.

```python
def countWord(filename, word):
    with open(filename, 'r') as f:
        text = f.read()
        text = text.lower()
        pos = text.find(word) # 인덱스
        count = 0
        while pos != -1:
            count +=1
            pos = text.find(word, pos+1) # string.find(찾을 문자, 시작 Indesm 끝 Index)
    return count

filename = input('filename을 입력하세요: ')
word = input('filename에서 개수를 구할 단어를 입력하세요: ')
ret = countWord(filename, word)
print(f'{word}의 개수: {ret}')
```

#### 파이썬에서 -1은 다양한 방식으로 사용됩니다. 다음은 일부 예시입니다:

- 인덱스: 파이썬에서는 리스트, 문자열, 튜플 등의 시퀀스 자료형의 인덱스를 지정할 때 음수를 사용할 수 있습니다. 이때 -1은 시퀀스의 마지막 요소를 가리킵니다. 예를 들어, 리스트에서 my_list[-1]은 리스트의 마지막 요소를 반환합니다.

- 슬라이스: 마찬가지로, 시퀀스 자료형의 슬라이스를 지정할 때도 음수를 사용할 수 있습니다. 이때 -1은 시퀀스의 마지막 요소를 가리킵니다. 예를 들어, 리스트에서 my_list[:-1]은 마지막 요소를 제외한 모든 요소를 반환합니다.

- 반환 값: 일부 함수에서는 실패 또는 오류를 나타내는 값을 -1로 반환합니다. 예를 들어, 문자열에서 find() 메서드는 문자열에서 지정된 문자열을 찾을 수 없을 때 -1을 반환합니다.

- 비트 연산: 비트 연산에서는 -1이 모든 비트가 1인 이진수로 사용됩니다. 예를 들어, 0b11111111은 10진수로 255이며, -1과 같습니다.
