#### 디테일
- 가계부 작성시 수입, 지출 등에서 문자 입력보다 숫자 입력이 편하기에 리스트가 아닌 딕셔너리로 구현해 보았음

- pickle() 사용: txt 파일에는 문자열 밖에 저장되지 않는 것 같아 시퀀스나 맵핑, 복합 자료형을 그대로 저장하고 불러올 수 있도록 pickle()을 사용함 (binary - packing, unpacking 방법)

#### 업데이트 예정 사항
- 조회시 금액 합계
- 분석 카테고리에 다양한 분석 하나씩 추가 예정
