```python
import pickle

category_income = {1:'급여', 2:'부캐', 3:'투자', 4:'용돈', 5:'당근'}
category_spending = {1:'식비', 2:'식자재', 3:'취미', 4:'핸드폰비', 5:'넷플릭스'}

with open('category_income.txt', 'wb') as f:
  pickle.dump(category_income, f)


with open('category_spending.txt', 'wb') as f:
  pickle.dump(category_spending, f)

```

```python
# # 개인비서 ❤👍👍 - 계산기 음식추천 카페추천 가계부 날씨 등

# 12/4 카테고리 감자 1000  (출력은 자동으로 format으로 , 바꾸어 주기)


# time() 사용
  # 가계부 사용자에게 도움이 되도록 시간을 나타내기 위해 사용

# 가계부 작성시 수입, 지출 등에서 문자 입력보다 숫자 입력이 편하기에 리스트가 아닌 딕셔너리로 구현해 보았음

# text 파일 사용
  # 프로그램 종료-재시작 후에도, 데이터와 새로 입력한 항목 등이 사라지지 않도록 txt 파일 사용

# pickle() 사용
   # txt 파일에는 문자열 밖에 저장되지 않는 것 같아 시퀀스나 맵핑, 복합 자료형을 그대로 저장하고 불러올 수 있도록 pickle()을 사용함 (binary - packing, unpacking 방법)

# try-except 예외처리 사용
  # input으로 문자 데이터를 받을시, int()관련 코드 에러를 넘기기 위해 사용
  # pickle.load()는 readline()처럼 한 줄씩 밖에 출력하지 않아, 전체 결과를 보려면 while()으로 돌려야하고, 끝 라인에서 아직 중단 방법을 찾지 못하여 'Ran out of input' 에러 발생함 -> 예외처리로 일단 넘김


import time
import pickle

t = time.localtime() # struct time sequence object를 사용하려면 변수 할당 뒤 변수.tm_year 이런식으로 사용해야 함 (아니면 time.localtime().tm_year)
week = ['월요일', '화요일','수요일','목요일','금요일','토요일','일요일']


print('안녕하세요. 박주경님, "부자되려면 가계부부터 써!!" 시작하겠습니다.\n')
print(f'오늘은 {t.tm_year}년 {t.tm_mon}월 {t.tm_mday}일 {week[t.tm_wday]} {t.tm_hour}시 {t.tm_min}분 {t.tm_sec}초입니다.\n')



while 1:
  print('1.수입', '2.지출', '3.검색 및 분석', '(종료: q)')
  i = input('원하시는 항목 번호를 선택하여 주세요> ')



# 수입 항목
  if i == '1':

    with open('category_income.txt', 'rb') as f:
      category_income = pickle.load(f)    

    while 1:
      print(category_income)
      i1 = input('원하시는 항목 번호를 선택해 주세요(항목 추가: a / 이전 단계: p)> ') # 'a', 'p' 등의 문자열이 먹어야 하기 때문에 int() X

      try: # 두 번째 elif에서 나올 수 있는 에러 처리('invalid literal for int() with base 10:') - base 10 = 십진법->정수
        if i1 == 'a':
          while 1:
            add1 = input('추가할 항목을 입력해 주세요(이전 단계: p)> ')
            if add1.strip() != "" and add1 != 'p' and add1 not in category_income.values(): # ""과 strip()을 써서 enter나 스페이스 입력 등에 대해 대처함, 'p'는 제외시킴
              for i in list(category_income.keys()): # dictionary 마지막에 추가하기 위해 for 반복문 이용
                i += 1
              category_income[i] = add1
              with open('category_income.txt','wb') as f: # 추가한 항목을 프로그램 재실행 때도 그대로 불러올 있도록 txt 파일로 저장 / # wb로 덮어쓰기해서 프로그램 재시작시 추가 항목 유지(ab로 하면 첫 번째 줄만 계속 불러옴, 추가시마다 라인 누적)
                pickle.dump(category_income, f)
              # with open('category_income.txt','rb') as f:
              #   while 1:
              #     print(pickle.load(f)) # 예외처리 때문에 Ran out of input도 넘어가는 듯
              print(category_income)
              amount1 = int(input('금액을 입력해 주세요> '))
              data = [[t.tm_year, t.tm_mon, t.tm_mday], add1, amount1, '수입']
              with open('financial_ledger.txt','ab') as f: #txt 파일에 정보 저장함 (추후 검색 및 분석 가능하도록)
                pickle.dump(data, f)
              print("\n기록했습니다.\n")
              break

            elif add1 in category_income.values():
              print('항목이 있습니다.')

            elif add1 == 'p':
              print('\n이전 단계로 돌아갑니다.\n')
              break

            else:
              print("잘못 입력하셨습니다.")
                      

        elif i1 == 'p':
          print('\n이전 단계로 돌아갑니다.\n')      # 예외 처리 때문에 'p' 작동 위해 아래 elif보다 위에 위치해야 함
          break



        elif int(i1) > 0 and int(i1) == list(category_income.keys())[int(i1)-1]: # 여기에 해당하는 숫자는 int여야 하기 때문에 int 처리
          while 1:
            try:
              amount2 = int(input('금액을 입력해 주세요> '))
              data = [[t.tm_year, t.tm_mon, t.tm_mday], list(category_income.values())[int(i1)-1], amount2, '수입']             # 금액 입력 부분에서 q p 등 처리해야 함
              with open('financial_ledger.txt','ab') as f:
                pickle.dump(data,f)
              print("\n기록했습니다.\n")
              # with open('financial_ledger.txt','rb') as f:
              #   while 1:              
              #     print(pickle.load(f))
              break
            except:
              print('잘못 입력하셨습니다.')

        else:
          print('잘못 입력하셨습니다.') # 0이나 음수, 카테고리 인덱스를 넘어가는 숫자, 이상한 숫자 등 처리

      except:
        print('잘못 입력하셨습니다.') # int() 정수 변경 시, 에러 나는 자료값들을 처리

    # break



# 지출 항목

  elif i == '2':
    
    with open('category_spending.txt', 'rb') as f:
      category_spending = pickle.load(f)


    while 1:
      print(category_spending)
      i2 = input('원하시는 항목 번호를 선택해 주세요(항목 추가: a / 이전 단계: p)> ')
      try:
        if i2 == 'a':
          while 1:
            add2 = input('추가할 항목을 입력해 주세요> ')
            if add2.strip() != "" and add2 != 'p' and add2 not in category_spending.values():
              for i in list(category_spending.keys()):
                i += 1
              category_spending[i] = add2
              with open('category_spending.txt','wb') as f:
                pickle.dump(category_spending, f)
              print(category_spending)
              amount3 = int(input('금액을 입력해 주세요> '))
              data = [[t.tm_year, t.tm_mon, t.tm_mday], add2, amount3, '지출']
              with open('financial_ledger.txt','ab') as f:
                pickle.dump(data, f)
              print("\n기록했습니다.\n")
              break


        elif i2 == 'p':
          print('\n이전 단계로 돌아갑니다.\n')
          break


        elif int(i2) > 0 and int(i2) == list(category_spending.keys())[int(i2)-1]:
          while 1:
            try:
              amount4 = int(input('금액을 입력해 주세요> '))
              data = [[t.tm_year, t.tm_mon, t.tm_mday], list(category_spending.values())[int(i2)-1], amount4, '지출']              # 금액 입력 부분에서 q p 등 처리해야 함
              with open('financial_ledger.txt','ab') as f:
                pickle.dump(data,f)
              print("\n기록했습니다.\n")
              # with open('financial_ledger.txt','rb') as f:              # 이 부분은 오류 발생 -> 뭔가 조치 더 안취하면 '금액 입력 무한 루프'
              #   while 1:
              #    print(pickle.load(f))
              break
            except:
              print("잘못 입력하셨습니다.")
              break
            
            # else:
            #   break


        # elif int(i2) > 0 and category_spending[int(i1)-2] in category_spending[:]:
          # print('있음')
          # break
        else:
          print('잘못 입력하셨습니다.')


      except:
        print('잘못 입력하셨습니다.')

    # break



# 검색 및 분석 항목

  elif i == '3':      

    while 1:
      print('1.수입 현황, 2.지출 현황, 3.내역별 검색, 4.날짜별 검색, (이전 단계: p)')
      i3 = input('원하시는 항목 번호를 선택해 주세요> ')

      try:
        if i3 == 'p':
          print('\n이전 단계로 돌아갑니다.\n')
          break



        elif int(i3) == 1:
          while 1:
            i3_income = input('"전체"나 원하는 지출 내역을 입력해 주세요> ')
            if i3_income.strip() == "":     # 공백, 스페이스 공백 거르기
              continue
            else:
              break            
          with open('financial_ledger.txt','rb') as f:
            try:
              while 1:
                data = pickle.load(f)
                if i3_income == '전체' and data[3] == '수입':
                  print(data)
                elif i3_income in data[1] and data[3] == '수입':
                  print(data)
                else:
                  continue
            except:
              print("\n(출력 완료)\n")



        elif int(i3) == 2:
          while 1:
            i3_spending = input('"전체"나 원하는 지출 내역을 입력해 주세요> ')
            if i3_spending.strip() == "" :      # 공백, 스페이스 공백 거르기
              continue
            else:
              break            
          with open('financial_ledger.txt','rb') as f:
            try:
              while 1:
                data = pickle.load(f)
                if i3_spending == '전체' and data[3] == '지출':
                  print(data)
                elif i3_spending in data[1] and data[3] == '지출':
                  print(data)
                else:
                  continue
            except:
              print("\n(출력 완료)\n")
          


        elif int(i3) == 3:
          while 1:
            i3_detail = input('검색할 내역을 입력해 주세요> ')
            if i3_detail.strip() == "" :      # 공백, 스페이스 공백 거르기
              continue
            else:
              break
          with open('financial_ledger.txt','rb') as f:
            try:
              while 1:
                data = pickle.load(f)
                if i3_detail in data[1]:
                  print(data)
            except:
              print("\n(출력 완료)\n")

              
# 날짜별 검색 
        elif int(i3) == 4:
          print('1.년도별 검색, 2.월별 검색, 3.일별 검색, (이전 단계: p)')
          i3_date = int(input('원하시는 항목 번호를 선택해 주세요> '))
       
          with open('financial_ledger.txt','rb') as f:
            try:
              if i3_date == 1:
                i3_year = int(input('년도 네 자리를 입력해 주세요> '))
                while 1:
                  data = pickle.load(f)
                  if i3_year == data[0][0]:
                    print(data)
                  else:
                    continue

              if i3_date == 2:
                i3_mon = input('달을 입력해 주세요> ')
                i3_mon = int(i3_mon.strip('0'))     # 월을 03 등으로 입력하면 0 없애기
                while 1:
                  data = pickle.load(f)
                  if i3_mon == data[0][1]:
                    print(data)
                  else:
                    continue


              if i3_date == 3:
                i3_day = input("해당 날짜를 '3/19'와 같이 입력해 주세요> ")
                i3_day = i3_day.lstrip('0')
                i3_day = i3_day.split('/')   # 0을 없애고 / 기준으로 리스트 만들기

                while 1:
                  data = pickle.load(f)
                  if int(i3_day[0]) == data[0][1] and int(i3_day[1]) == data[0][2]:
                    print(data)
                  else:
                    continue


            except:
              print("\n(출력 완료)\n")



        else:                       # int값 처리
          print('잘못 입력하셨습니다.')



      except:                       # 문자열 등 처리
        print("잘못 입력하셨습니다.")


      # 날짜 - 세부내역과 총합
      # 카테고리 - 카테고리별 세부내역과 총합
      # 분석 max
      # 월별 카테고리 비교 - 지출의 증감
      # 카테고리별 금액



  elif i == 'q':

    print("종료합니다.")
    break

    



  else:
    continue
```
