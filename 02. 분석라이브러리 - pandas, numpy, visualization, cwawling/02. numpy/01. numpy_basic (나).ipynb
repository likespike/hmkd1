- 정방행렬은 행과 열의 수가 같은 행렬 (행렬방정식의 기본이 되는 행렬. 차수가 n)
- 대각행렬은 주대각선 원소를 제외한 모든 원소들이 0인 정방행렬
- 삼각행렬은 주대각선 원소를 기준으로 위 또는 아래에 있는 성분이 모두 0인 정방행렬
- 항등행렬은 행렬 곱셈 연산에 항등원으로 작용하는 행렬 (주대각선 성분이 모두1인 정방행렬) (= 단위 행렬)
- 영행렬은 모든 원소가 0인 행렬로 곱셈 연산에서 영원으로 작용하는 행렬
- 전치행렬은 주대각선 원소를 기준으로 행과 열을 바꿔주는 행렬
- 직교행렬은 행렬 A의 역행렬이 A의 전치행렬이고 A의 전치행렬과 A 행렬을 곱하였을때 항등행렬이 나오는 행렬

#### # 직교행렬
- 행과 열이 서로 직교하는 정방행렬
- 모든 열벡터와 행벡터가 서로 직교하고 크기가 1인 단위백터로 이루어짐
- np.linalg.qr() 함수를 사용하여 QR 분해를 수행하여 직교행렬을 추출
- q를 추출하여 직교행렬 orth_mat을 만든다
- np.dot(orth_mat, orth_mat.T)를 계산하여 직교성을 검증
- np.allclose() 함수로 두 행렬이 동일한지 검사. True를 반환하면 두 행렬은 동일

```python
# 직교행렬은 행렬 A의 역행렬이 A의 전치행렬이고 A의 전치행렬과 A 행렬을 곱하였을때 항등행렬이 나오는 행렬

mat = np.random.randn(3,3)
print(mat,'\n')

# qr 분해
q,r = np.linalg.qr(mat)

# 직교행렬 추출
orth_mat = q
print(orth_mat,'\n')

# 직교성 검증
print(np.allclose(np.dot(orth_mat, orth_mat.T), np.eye(3)))    # np.dot(orth_mat, orth_mat.T)과 np.eye(3)이 동일한지 검증
```
