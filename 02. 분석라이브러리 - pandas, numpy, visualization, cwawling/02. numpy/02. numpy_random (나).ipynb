[Random 함수]  

- np.random.seed       # seed를 통한 난수 생성
- np.random.randint    # 균일분포의 정수 난수 1개 생성
- np.random.rand       # 0부터 1사이의 균일분포에서 난수 매트릭스 array 생성
    - 균일분포(uniform distribution)는 확률 분포의 하나로, 모든 값들이 동일한 확률로 나타날 때 이루어지는 분포입니다.예를 들어, 주사위를 던지는 경우를 생각해 봅시다. 주사위의 각 면은 나올 확률이 동일하기 때문에, 1에서 6까지의 모든 값들이 동일한 확률로 나타납니다. 이 때, 나타나는 값들의 분포를 균일분포라고 합니다.
- np.random.randn      # 가우시안 표준 정규 분포에서 난수 매트릭스 array 생성
- np.random.shuffle    # 기존의 데이터의 순서 바꾸기
- np.random.choice     # 기존의 데이터에서 sampling

- gender : 0,1 정수 난수 생성

```python
import numpy as np
import pandas as pd

ar = np.zeros((1000,15), dtype=int)
ar[:,0] = np.arange(1,1001)
ar[:,1] = np.random.randint(2, size=1000)
ar[:,2] = np.random.randint(1, 81, size=1000)
ar[:,3] = np.random.randint(1, 11, 1000)
ar[:,4] = np.random.choice([11111,22222,33333,44444,55555], 1000)
ar[:,5] = np.random.choice([11111,22222,33333,44444,55555], 1000)
ar[:,6] = np.random.randint(1, 51, 1000) * 1000
ar[:,7] = np.random.randint(1, 51, 1000) * 1000
ar[:,8] = np.random.randint(1, 101, 1000)
ar[:,9] = np.random.randint(1, 101, 1000)
ar[:,10] = np.random.randint(1, 24, 1000)
ar[:,11] = np.random.randint(1, 24, 1000)
ar[:,12] = ar[:,6] * ar[:,8]
ar[:,13] = ar[:,7] * ar[:,9]
ar[:,14] = np.where(((ar[:,13] - ar[:,12]) > 0), 1, 0)

columns = ['id','gender','age','region','prod_22', 'prod_23', 'price_22', 'price_23', 'qty_22', 'qty_23',
           'time_22', 'time_23', 'amount_22', 'amount_23', 'sales']

df = pd.DataFrame(ar, columns=columns)

df.head()
```

```python
# 나

import numpy as np
import pandas as pd

np.random.seed(0)
ar = np.random.rand(1000, 15)

# 배열을 df로 변환
columns = ['id','gender','age','region','prod_22', 'prod_23', 'price_22', 'price_23', 'qty_22', 'qty_23',
           'time_22', 'time_23', 'amount_22', 'amount_23', 'sales']
df = pd.DataFrame(ar, columns = columns)


df.id = np.arange(1, 1001)
df.gender = np.random.randint(2, 1000)
df.age = np.random.randint(10, 81, 1000)
df.region = np.random.randint(1, 11, 1000)

df.prod_22 = np.random.choice([11111,22222, 33333, 44444, 55555], 1000)
df.prod_23 = np.random.choice([11111,22222, 33333, 44444, 55555], 1000)

# df.price_22 = np.random.randint(1, 50001, 1000).astype(float) # 이렇게 하면 안됨, 가격이니까!
df.price_22 = np.random.randint(1, 51, 1000).astype(float) * 1000 
df.price_23 = np.random.randint(11, 51, 1000).astype(float) * 1000 
df.qty_22 = np.random.randint(1, 101, 1000)
df.qty_23 = np.random.randint(1, 101, 1000)

df.time_22 = np.random.randint(1, 25, 1000) # 0붙이기(1)
df.time_22 = df.time_22.astype(str).str.zfill(2)

df.time_23 = np.random.randint(1, 25, 1000) # 0붙이기(2)
df.time_23 = np.where(df.time_23 < 10, '0'+df.time_23.astype(str), df.time_23)

df.amount_22 = df.price_22 * df.qty_22
df.amount_23 = df.price_23 * df.qty_23

df.sales = df.amount_23 - df.amount_22
df.sales = np.where(df.sales > 0, 1, 0)

df
```
