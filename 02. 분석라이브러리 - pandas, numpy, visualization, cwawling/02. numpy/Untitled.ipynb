```python
import numpy as np
import pandas as pd

ar = np.zeros((1000,17), dtype=int)

ar[:,0] = np.arange(1,1001)
ar[:,1] = np.random.randint(2, size=1000)    # 0 = 남자, 1 = 여자
ar[:,2] = np.random.randint(10, 91, size=1000)
ar[:,3] = np.random.randint(1, 11, 1000)
ar[:,4] = np.random.choice([11, 22, 33, 44, 55], 1000)
ar[:,5] = np.random.choice([11, 22, 33, 44, 55], 1000)
ar[:,6] = np.random.randint(1, 51, 1000) * 1000
ar[:,7] = np.random.randint(1, 51, 1000) * 1000
ar[:,8] = np.random.randint(1, 101, 1000)
ar[:,9] = np.random.randint(1, 101, 1000)
ar[:,10] = np.random.randint(1, 24, 1000)
ar[:,11] = np.random.randint(1, 24, 1000)
ar[:,12] = np.random.randint(1, 6, 1000)
ar[:,13] = np.random.randint(1, 11, 1000)
ar[:,14] = ar[:,6] * ar[:,8]
ar[:,15] = ar[:,7] * ar[:,9]
ar[:,16] = np.where(((ar[:,15] - ar[:,14]) > 0), 1, 0)

columns = ['id','gender','age','region','pdt_22', 'pdt_23', 'pa_22', 'pa_23', 'q_22', 'q_23',
           't_22', 't_23', 'channel', 'tr_y', 'pur_19', 'pur_20', 'sales']

df = pd.DataFrame(ar, columns=columns)


# 각 컬럼별로 10개씩 null 값을 삽입
for i in range(1,17):
    idx = np.random.choice(df.index, size=1, replace=False)
    df.iloc[idx, i] = np.nan

    
df
```

```python
# gender가 1인 index 추출
idx = df[df['gender'] == 1].index

# pdt_22의 값 중 70%를 11.0으로 바꿔줌
df.loc[np.random.choice(idx, size=int(len(idx)*0.7), replace=False), 'pdt_22'] = 11.0
```
