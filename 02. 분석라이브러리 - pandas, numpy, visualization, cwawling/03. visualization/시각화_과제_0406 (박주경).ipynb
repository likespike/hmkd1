```python
# relplot -> matplotlib

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors

# 색상 이름 매핑을 위한 딕셔너리 생성
color_dict = dict(zip(data_kj['Nationality'].unique(), mcolors.CSS4_COLORS.keys()))

# plt.scatter로 그리기
plt.scatter(data=data_kj, x='Overall', y='Nwage', c=data_kj['Nationality'].map(color_dict), alpha=1)
plt.xlabel('Overall')
plt.ylabel('Nwage')
plt.show()
```

```python
# relplot -> plot

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# 데이터 프레임 생성
df = pd.DataFrame({'Overall': data_kj['Overall'], 'Nwage': data_kj['Nwage'], 'Nationality': data_kj['Nationality']})

# 색상 이름 매핑을 위한 딕셔너리 생성
color_dict = dict(zip(df['Nationality'].unique(), mcolors.CSS4_COLORS.keys()))

# plot 메소드로 그리기
fig, ax = plt.subplots()
for key, group in df.groupby('Nationality'):
    group.plot(ax=ax, kind='scatter', x='Overall', y='Nwage', label=key, color=color_dict[key], alpha=1)

plt.xlabel('Overall')
plt.ylabel('Nwage')
plt.show()
```
