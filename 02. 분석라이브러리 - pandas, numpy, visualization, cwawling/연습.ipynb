```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# "ticket" 열의 값이 "1"로 시작하는 데이터 추출
ticket_starts_with_1 = df[df['ticket'].str.startswith('69')]

# "pclass"와 "ticket" 열 값이 "1"로 시작하는 티켓의 관계 시각화
sns.countplot(data=ticket_starts_with_1, x='pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Relationship between Pclass and Tickets Starting with 1')
plt.show()
```
